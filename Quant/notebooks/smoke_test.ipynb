```python
import yfinance as yf
import vectorbt as vbt
import pandas as pd
```

```python
# 1. Download 2 years of daily Bitcoin data
print("Fetching market data...")
data = yf.download("BTC-USD", start="2024-01-01", end="2026-01-01")
```

```python
# 2. Extract the Closing Price
close_prices = data['Close']
```

```python
# 3. Define a quick technical strategy: Buy when 10-day moving average > 50-day moving average
fast_ma = vbt.MA.run(close_prices, 10)
slow_ma = vbt.MA.run(close_prices, 50)
entries = fast_ma.ma_crossed_above(slow_ma)
exits = fast_ma.ma_crossed_below(slow_ma)
```

```python
# 4. Run a vectorized backtest
print("Simulating trading strategy...")
portfolio = vbt.Portfolio.from_signals(close_prices, entries, exits, fees=0.001)
```

```python
# 5. Output the results
print("\n--- Strategy Performance Summary ---")
print(f"Total Return: {portfolio.total_return().values[0]*100:.2f}%")
print(f"Sharpe Ratio: {portfolio.sharpe_ratio().values[0]:.2f}")
```
