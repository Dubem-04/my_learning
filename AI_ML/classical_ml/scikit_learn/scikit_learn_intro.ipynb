**What is scikit-learn?**

Scikit-learn is Python's most widely used machine learning library. Built on top of NumPy, SciPy, and Matplotlib, it provides a clean, consistent interface for the entire supervised and unsupervised ML workflow.

---

**The Core API - one pattern, every model**

The single most important thing to understand is that *every* estimator in scikit-learn follows the same three-method contract:

- `fit(X_train, y_train)` — train the model on data
- `predict(X_test)` — generate predictions on new data
- `score(X_test, y_test)` — return a default metric (accuracy for classifiers, R² for regressors)

This uniformity means swapping a Logistic Regression for a Random Forest is a one-line change. It also makes pipelines and cross-validation work seamlessly across all models.

---

**Algorithm families**

Scikit-learn covers three major ML paradigms:

- **Supervised — Classification**: Logistic Regression, SVMs, K-Nearest Neighbors, Decision Trees, Random Forest, Gradient Boosting, Naive Bayes
- **Supervised — Regression**: Linear/Ridge/Lasso Regression, SVR, Random Forest Regressor, Gradient Boosting Regressor
- **Unsupervised**: K-Means, DBSCAN (clustering); PCA, t-SNE (dimensionality reduction); Isolation Forest (anomaly detection)

---

**Preprocessing (`sklearn.preprocessing`)**

Before any model sees data, scikit-learn handles:

- Feature scaling: `StandardScaler`, `MinMaxScaler`, `RobustScaler`
- Encoding categoricals: `OneHotEncoder`, `OrdinalEncoder`, `LabelEncoder`
- Imputation of missing values: `SimpleImputer`, `KNNImputer`
- Polynomial and interaction features: `PolynomialFeatures`

---

**Pipelines (`sklearn.pipeline`) - the real superpower**

`Pipeline` chains preprocessing and modeling steps into a single object. This is critical because it:

1. Eliminates data leakage (preprocessing is fit only on training data, then applied to test data)
2. Makes cross-validation and hyperparameter tuning clean and correct
3. Lets you deploy a single object that handles both transformation and prediction

`ColumnTransformer` extends this to apply different transformers to different columns simultaneously.

---

**Model selection & tuning (`sklearn.model_selection`)**

- `train_test_split`: split data into train/test sets
- `cross_val_score`: k-fold cross-validation in one call
- `GridSearchCV`: exhaustive hyperparameter search over a parameter grid
- `RandomizedSearchCV`: faster sampling-based search for large parameter spaces

---

**Evaluation metrics (`sklearn.metrics`)**

Scikit-learn ships a full library of metrics:

| Task | Metrics |
|---|---|
| Classification | accuracy, precision, recall, F1, ROC-AUC, confusion matrix |
| Regression | MSE, RMSE, MAE, R² |
| Clustering | Silhouette score, Davies-Bouldin index |

---

**Utilities**

- **Built-in toy datasets**: `load_iris()`, `load_digits()`, `fetch_california_housing()` - great for learning and benchmarking
- **Feature selection**: `SelectKBest`, `RFE` (Recursive Feature Elimination)
- **Model persistence**: save/load models with `joblib` or `pickle`

---

**What scikit-learn is not for?**

It doesn't handle deep learning (use PyTorch or TensorFlow for that), large-scale distributed training (use Spark MLlib or Dask-ML), or sequential/time-series-specific models beyond basics. It's the gold standard for classical ML on tabular data.
