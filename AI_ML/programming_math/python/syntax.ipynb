### **Python Variables**

Variables are containers for storing data values

In Python, a variable is created once a value is assigned to it using the assignment operator `=`

Rules for naming variables:
- A variable name must start with a letter or the underscore character.
- A variable name cannot start with a number.
- A variable name can only contain alpha-numeric characters and underscores (A-z, 0-9, and _ ).
- Variable names are case-sensitive (age, Age and AGE are three different variables).
- A variable name cannot be any of the Python keywords.

##### **Multi Words Variable Names**

Each word, except the first, starts with a capital letter:

```python
myVariableName = "John" # Camel case
```

Each word starts with a capital letter:

```python
MyVariableName = "John" # Pascal case
```

Each word is separated by an underscore character:

```python
my_variable_name = "John" # Snake case
```

##### **Unpack a Collection**

Unpacking is a convenient and readable feature that allows you to extract values from iterables (like lists, tuples, or strings) and assign them to multiple variables in a single line of code.

**Basic Unpacking**

```python
fruits = ["apple", "banana", "cherry"]
x, y, z = fruits
print(x, y, z, sep=("\n"))
```

**Extended unpacking**

If you have an iterable with many items but you only care about a few specific ones, you can use the asterisk (`*`) to "catch" the remaining items into a new list. This is called extended iterable unpacking.

```python
grades = [95, 85, 92, 78, 99]

# Grab the first grade and pack the rest into a list
first, *rest = grades
print([first], rest)

# It can also be used to grab the last item
*previous, last = grades
print(previous, [last])

# Or grab the first and last while packing the middle
first, *middle, last = grades
print([first], middle, [last])
```

**Dictionary Unpacking**

While the single asterisk (`*`) unpacks the keys of a dictionary, the double asterisk (`**`) unpacks the *key-value* pairs.

```python
dict1 = {
    "a": 1,
    "b": 2
}
dict2 = dict(c = 3, d = 4)

print("Unpacked keys:", *dict1, *dict2)

merged = {**dict1, **dict2}
print("Unpacked and merged key-value:", merged)
```

*Note: If there are overlapping keys, the value from the dictionary unpacked last will overwrite the earlier one.*

**Unpacking in Function Arguments**

```python
def print_user(name, age, city):
    print(f"{name} is {age} and lives in {city}")

user_list = ["Bob", 30, "New York"]
user_dict = {"name": "Charlie", "age": 28, "city": "London"}

# Unpacking a list into positional arguments
print_user(*user_list)

# Unpacking a dictionary into keyword arguments
print_user(**user_dict)
```

#### **Global Variables**

These are variables created outside of a function (as in all of the examples above) are known as global variables. These variables can be used by everyone, both inside of functions and outside.

```python
x = "awesome"

def my_func():
    print("Python is " + x)

my_func()
```

If you create a variable with the same name inside a function, this variable will be local, and can only be used inside the function. The global variable with the same name will remain as it was (global and with the original value).

```python
def my_func():
    x = "fantastic"
    print("Python is " + x)

my_func()

print("Python is " + x)
```

Normally, when you create a variable inside a function, that variable is local, and can only be used inside that function. To create a global variable inside a function, you can use the `global` keyword.

```python
def my_func():
    global x
    x = "fantastic"
    print("Python is " + x)

my_func()

print("Python is " + x)
```
